```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Knowledge base (simulated documents)
documents = [
    "DevelopersHub Corporation was founded in 2018 and specializes in AI/ML solutions.",
    "The AI/ML internship program runs for 3 months with hands-on projects.",
    "Interns work on real datasets including Iris, Telco Churn, and House Prices.",
    "The internship deadline for Task 1 is 26th June 2026.",
    "Task 2 involves building ML pipelines with scikit-learn.",
    "Task 3 requires heart disease prediction using logistic regression.",
    "Task 4 is about building chatbots with RAG and LangChain.",
    "Task 5 uses LLMs for auto-tagging support tickets.",
    "Task 6 focuses on house price prediction with gradient boosting.",
    "The company office is located in San Francisco, California.",
    "DevelopersHub offers full-time positions to top-performing interns.",
    "The tech stack includes Python, TensorFlow, PyTorch, and Hugging Face."
]

# Simple retrieval function
vectorizer = TfidfVectorizer()
doc_vectors = vectorizer.fit_transform(documents)

def retrieve_docs(query, top_k=2):
    query_vec = vectorizer.transform([query])
    similarities = cosine_similarity(query_vec, doc_vectors).flatten()
    top_indices = similarities.argsort()[-top_k:][::-1]
    return [documents[i] for i in top_indices], similarities[top_indices]

# Test retrieval
test_query = "What is the internship deadline?"
retrieved, scores = retrieve_docs(test_query)
print(f"Query: {test_query}")
print(f"Retrieved docs (scores: {scores}):")
for doc in retrieved:
    print(f"  - {doc}")
```

```python
class SimpleRAGChatbot:
    def __init__(self):
        self.conversation_history = []
    
    def chat(self, user_input):
        # Retrieve relevant documents
        retrieved_docs, _ = retrieve_docs(user_input, top_k=2)
        context = " ".join(retrieved_docs)
        
        # Simple response generation (rule-based)
        response = self.generate_response(user_input, context)
        
        # Store in memory
        self.conversation_history.append({"user": user_input, "bot": response})
        
        return response
    
    def generate_response(self, query, context):
        query_lower = query.lower()
        
        # Simple intent matching
        if "deadline" in query_lower or "due date" in query_lower:
            return f"Based on our records: {context} The specific deadline information is available in the internship guide."
        
        elif "task" in query_lower and "what" in query_lower:
            return f"Here's what I found: {context} Would you like more details about any specific task?"
        
        elif "company" in query_lower or "about" in query_lower:
            return f"DevelopersHub info: {context}"
        
        elif "location" in query_lower or "office" in query_lower:
            return f"Location details: {context}"
        
        elif "internship" in query_lower or "program" in query_lower:
            return f"Internship program details: {context}"
        
        else:
            return f"I found this relevant information: {context} For more specific questions, please check the official documentation."
    
    def get_history(self):
        return self.conversation_history

# Test the chatbot
bot = SimpleRAGChatbot()

conversations = [
    "What is the internship deadline?",
    "Tell me about Task 2",
    "Where is the company located?",
    "What tasks do interns work on?",
    "Does the company offer full-time jobs?"
]

print("=" * 60)
for query in conversations:
    response = bot.chat(query)
    print(f"User: {query}")
    print(f"Bot: {response}")
    print("=" * 60)

print("\nConversation History:")
for i, turn in enumerate(bot.get_history(), 1):
    print(f"{i}. User: {turn['user']}")
    print(f"   Bot: {turn['bot'][:80]}...")
```
